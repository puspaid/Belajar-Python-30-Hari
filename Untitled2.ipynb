```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 1. Mengambil Dataset langsung dari GitHub
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv(url, names=names)

# 2. Pemisahan X (Fitur) dan y (Target)
X = df.drop('class', axis=1)
y = df['class']

# 3. Scaling (Penting karena rentang angka tiap kolom berbeda jauh)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Bagi data (80% Training, 20% Testing)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 5. Siapkan List Model
models = {
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "Naive Bayes": GaussianNB()
}

# 6. Melatih Model dan Menghitung Metrik
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred)
    })

# 7. Tampilkan Tabel Ringkasan
df_hasil = pd.DataFrame(results)
print("--- Ringkasan Hasil Evaluasi Model ---")
print(df_hasil.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# 1. Gaskan buat ulang model Decision Tree khusus untuk visualisasi
# Kita batasi max_depth=3 agar pohonnya tidak terlalu rumit dilihat
model_viz = DecisionTreeClassifier(max_depth=3, random_state=42)
model_viz.fit(X_train, y_train)

# 2. Atur ukuran gambar
plt.figure(figsize=(20,10))

# 3. Gambar Pohonnya
tree.plot_tree(model_viz,
               feature_names=names[:-1],  # Nama fitur (preg, plas, dsb)
               class_names=['Sehat', 'Diabetes'], # Nama kategori
               filled=True,
               rounded=True,
               fontsize=12)

plt.title("Aturan 'Jika-Maka' pada Decision Tree (Prediksi Diabetes)")
plt.show()
```
